```
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import norm
domain=np.linspace(-2,2,1000)
plt.plot(domain,norm.pdf(domain,0,1))
plt.title("standard deviation")
plt.xlabel("value")
plt.yabel("density")
plt.show()
```
